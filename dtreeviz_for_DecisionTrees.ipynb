```python
from sklearn.datasets import load_iris
from dtreeviz.trees import dtreeviz, tree
```

```python
iris_data = load_iris()

X_train = iris_data.data
y_train = iris_data.target


model = tree.DecisionTreeClassifier()

model.fit(X_train, y_train)
```

```python
# Basic plot tree method
from sklearn.tree import plot_tree
plot_tree(model)
```

```python
# using dtreeviz
viz = dtreeviz(model,
               X_train,
               y_train,
               feature_names = iris_data.feature_names)
viz
```

```python
# display dtreeviz horizontally
viz = dtreeviz(model,
               X_train,
               y_train,
               feature_names = iris_data.feature_names,
               orientation='LR' )
viz
```

```python
# dotted highlighed data point shows the path data point is following
viz = dtreeviz(model,
               X_train,
               y_train,
               feature_names = iris_data.feature_names,
               class_names = list(iris_data.target_names),
               X=X_train[124])
viz
```
